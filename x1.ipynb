```python

# 결측치 개수 & 비율 계산
missing_count = df.isnull().sum()
missing_ratio = (missing_count / len(df)) * 100

missing_df = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing Ratio (%)": missing_ratio
}).sort_values(by="Missing Ratio (%)", ascending=False)

print("✅ 결측치 비율 확인 완료")
missing_df.head(20)  # 상위 20개 컬럼만 보기
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
missing_df[missing_df["Missing Ratio (%)"]>0].sort_values(by="Missing Ratio (%)")["Missing Ratio (%)"].plot(kind="barh", color="salmon")
plt.title("결측치 비율(%) 상위 변수")
plt.xlabel("Missing Ratio (%)")
plt.ylabel("Columns")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 수치형 변수만 선택
num_cols = ['price', 'shipping_charges', 'volume_cm3', 'weight_kg',
            'approval_delay_minutes', 'delivery_delay_days']

corr_matrix = df[num_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("수치형 변수 상관관계 Heatmap")
plt.show()

```
