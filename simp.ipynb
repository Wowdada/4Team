```python
def get_ttest(df):
    from scipy.stats import ttest_ind
    
    # 컬럼명 공백 제거
    df.columns = df.columns.str.strip()
    
    # 결측치 제거
    male = df[(df['gender'] == '남성') & (df['review_score'].notna())]['review_score']
    female = df[(df['gender'] == '여성') & (df['review_score'].notna())]['review_score']
    
    # t-test 실행
    t_stat, p_value = ttest_ind(male, female, equal_var=True)
    
    return round(t_stat, 4), round(p_value, 4)

# 실행
print(get_ttest(df))
```
